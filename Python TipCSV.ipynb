# Python Tip Data Project

**This project takes a look at and briefly analyzes tip data at a restaurant.  The initial data provided includes information on the total bill, tip given, gender of the customer, the day of the week, time of day, size of the party, and whether the customer is a smoker or not.**

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# read the data from the tips.csv file
tip = pd.read_csv("tips.csv")
```

**Looking at the column headers and first several rows of data:**

```python
tip.head()
```

## Tip Percentage

**Calculating the percentage of the tip to the whole bill and the overall average and median tips and tip percentage.**

```python
#calculate tip percentage of total bill
tipperc = tip["tip"] / tip["total_bill"] *100

#round to 2 decimal places
roundtip=round(tipperc, 2)

#make a new column to display tip percentage
tip["TipPercent"]=roundtip

#call the first five rows to check
tip.head()
```

```python
# find the mean bill in dollars and cents
print("The mean bill is $",round(tip["total_bill"].mean(),2))

# find the median bill in dollars and cents
print("The median bill is $",round(tip["total_bill"].median(),2))

print(" ")

# find the mean tip in dollars and cents
print("The mean tip is $",round(tip["tip"].mean(),2))

# find the median tip in dollars and cents
print("The median tip is $",round(tip["tip"].median(),2))

print(" ")

# find the mean tip percent to two decimal places
print("The mean tip percent is ",round(tip["TipPercent"].mean(),2),"%")

# find the median tip in dollars and cents
print("The median tip percent is ",round(tip["TipPercent"].median(),2),"%")
```

**<span style='color:Red'> Analysis: The mean tip and tip percent are very close to each other 10c and 1.05% respectively.  Although there may be outliers in the data it will be fairly evenly distributed. </span>**

## Male vs Female Tips
**A brief analysis of tips for male diners as compared to female diners.**

```python
#male tips
sub1 = tip[(tip['sex'] =='Male')]

x = sub1["tip"]
y = sub1["TipPercent"]

# axis size, labels, and graph title
plt.figure(figsize=(8, 8))
plt.xlabel('Tip Amount')
plt.ylabel('Tip Percentage of Bill')
plt.title('Male Tip Amount vs. Tip Percentage')

# render the scatter plot
plt.scatter(x,y)
plt.show()

#female tips
sub2 = tip[(tip['sex'] =='Female')]

x = sub2["tip"]
y = sub2["TipPercent"]

# axis size, labels, and graph title
plt.figure(figsize=(8, 8))
plt.xlabel('Tip Amount')
plt.ylabel('Tip Percentage of Bill')
plt.title('Female Tip Amount vs. Tip Percentage')

# render the scatter plot
plt.scatter(x,y)
plt.show()
```

**<span style='color:Red'> Analysis: The data is closely gathered and linear, but there is a significant outlier in the male data (70%) and in the female data (140%) as well as a few more modest anomalies in both sets that may skew averages. </span>**

### Male vs Female Tips Plot Comparison

```python
# data
subsetA = tip[(tip['sex'] =='Male')]
subsetB = tip[(tip['sex'] =='Female')]
subsetC = subsetA['tip']
subsetD = subsetB['tip']
data = list([subsetC, subsetD])

fig, ax = plt.subplots(figsize=(7, 7))

# build a box plot
ax.boxplot(data)

# title and axis labels
ax.set_title('Male vs Female Tips')
ax.set_xlabel('Diners')
ax.set_ylabel('Tip Amount')
xticklabels=['Male', 'Female']
ax.set_xticklabels(xticklabels)

# add horizontal grid lines
ax.yaxis.grid(True)

# show the plot
plt.show()
```

**<span style='color:Red'> Analysis: The minimum and median tip amounts are similar for both male and females, but the maximum tip amount and upper quartiles are higher for men as well as many more male outliers existing above the maximum.  Although as outliers go, the largest tip amount was left by a female. </span>**

## Party Size and Tips
**A brief analysis of how party size affects tips**

```python
import pandas as pd
from matplotlib.pyplot import pie, axis, show
%matplotlib inline

#chart for tip total by party size
fig, ax = plt.subplots(figsize=(10, 10))
sums = tip.groupby(tip["size"])["tip"].sum()
axis('equal')
pie(sums, labels=sums.index, autopct='%1.f%%', startangle=90)
ax.set_title("Total Tip Amounts by Party Size")
show()

#chart for average tip percentage by party size
fig, ax = plt.subplots(figsize=(6,6))
sums = tip.groupby(tip["size"])["TipPercent"].mean()
axis('equal')
pie(sums, labels=sums.index, autopct='%1.1f%%', startangle=90)
ax.set_title("Average Tip Percentage by Party Size")
show()
```

**<span style='color:Red'> Analysis: As the pie chart showing party size with percentage of total tips and average percentage of tips by party size (which is more evenly distributed) are vastly different, there is likely to be far more parties of 2 than parties of 1.  However, parties of 1, and to a lesser extent 2, appear to be more generous with tips (higher average percentage) than the larger groups (closer to 15% for groups of 3 or more)</span>**

## Day of the Week and Tips
**A brief analysis of how tips change over the course of the week**

```python
#plot for Total tips each day

#size of plot
plt.figure(figsize=(6,5))

#plot labels
plt.xlabel('Days of the Week')
plt.ylabel('Tips')
plt.title('Total Tips by Day of the Week')

#totaling days tips and determining axis
tip['Tiptotal']=tip.groupby('day')['tip'].transform('sum')
plt.bar(x=tip['day'], height=tip['Tiptotal'])

plt.show()

#plot for Average Tips each day

#size of plot
plt.figure(figsize=(6,5))

#plot labels
plt.xlabel('Days of the Week')
plt.ylabel('Tips')
plt.title('Average Tips by Day of the Week')

#totaling days tips and determining axis
tip['Tiptotal']=tip.groupby('day')['tip'].transform('mean')
plt.bar(x=tip['day'], height=tip['Tiptotal'])

plt.show()
```

**<span style='color:Red'> Analysis: Although total tips on Friday are lower on average than other days, it is within 50 cents of the highest day (Sunday).  Thus, the low tip total shown for Friday implies that there is less business done on Fridays than other days.  Additionally, tip totals are much higher on the weekend than on weekdays. </span>**

```python
#plot for Average Tip Percents each day

#size of plot
plt.figure(figsize=(6,5))

#plot labels
plt.xlabel('Days of the Week')
plt.ylabel('Tip Percent')
plt.title('Average Tip Percent by Day of the Week')

#totaling days tips and determining axis
tip['Tiptotal']=tip.groupby('day')['TipPercent'].transform('mean')
plt.bar(x=tip['day'], height=tip['Tiptotal'])

plt.show()
```

**<span style='color:Red'> Analysis: Although less business is done on weekdays, there appears to be a tendency toward higher tip percentages on the weekdays as opposed to the weekend. </span>**

## Size of the Bill and Tips
**A brief analysis of how the size of the bill affects tips**

However, when figuring the tip as a percentage of the bill against amount, there is a slight negative linear relationship.  This implies that as the bill total goes up, diners are slightly more likely to tip a lower percentage. </span>**
